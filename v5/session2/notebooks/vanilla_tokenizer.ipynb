```python
from tokenizers import Tokenizer, models, pre_tokenizers, trainers
```

```python
tokenizer = Tokenizer(models.BPE(unk_token="[UNK]"))
```

```python
tokenizer.pre_tokenizer = pre_tokenizers.Whitespace()
```

```python
trainer = trainers.BpeTrainer(
    vocab_size=300,
    min_frequency=2,
    special_tokens=["[UNK]"],
)
```

```python
def data_iterator():
    # Replace this with your actual text stream or dataset
    dataset = [
        "Transformers are excellent for language modeling.",
        "Hugging Face provides fast byte pair encoding tokenizers.",
        "Training a tokenizer from scratch is straightforward.",
    ]
    for text in dataset:
        yield text
```

```python
tokenizer.train_from_iterator(data_iterator(), trainer=trainer)
```

```python
tokenizer.save("vanilla_tokenizer.json")
```
